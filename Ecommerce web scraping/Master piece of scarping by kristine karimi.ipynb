```python
import requests
from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from random import randint
```

```python
# Loop to go over all pages
pages = np.arange(1, 25, 1)
data=[]
for page in pages:
    
    page="https://www.olx.ro/servicii-afaceri-colaborari/meseriasi-constructori/instalatii-sanitare-termice/?page/" + str(page) 
    sleep(randint(2,10))
    response = requests.get(page)
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    my_table = soup.find_all('a')
    for tag in my_table:
        data.append(tag.get('href'))
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
```

```python
data1=[]
for link in mega:
    f = requests.get(link,headers=headers).text
    hun=BeautifulSoup(f,'html.parser')

    try:
        price=hun.find("div",{"class":"css-g5mtbi-Text"}).text.replace('\n',"")
    except:
        price = None

    try:
        about=hun.find("h2",{'class':'css-u8mbra-Text eu5v0x0'}).text.replace('\n',"")
    except:
        about=None

    try:
        rating = hun.find("h2",{'class':'css-u8mbra-Text eu5v0x0'}).text.replace('\n',"")
    except:
        rating=None

    try:
        name=hun.find("h1").text.replace('\n',"")
    except:
        name=None

    whisky = {"Title":name,"desc":price,"Author":rating,"author":about}

    data1.append(whisky)
```

```python
import json
json_object = json.dumps(data1, indent = 4)
with open("SERVICE1.json", "w") as outfile:
    outfile.write(json_object)
```
